```python
from gssutils import *

if is_interactive():
    scraper = Scraper(
        "https://www.justice-ni.gov.uk/publications/"
        "research-and-statistical-bulletin-82017-views-alcohol-and-drug-related-issues-findings-october-2016")
scraper
```

```python
if is_interactive():
    tabs = {tab.name: tab for tab in scraper.distribution(
        title='October 2016 alcohol and drugs findings data table').as_databaker()}
tabs.keys()
```

https://www.justice-ni.gov.uk/sites/default/files/publications/justice/ni-omnibus-survey-oct-2016-alcohol-drugs-tabular.xlsx

```python
tab = tabs['Table A2']
```

```python
observations = tab.excel_ref('B5').expand(DOWN).expand(RIGHT).is_not_blank()
```

```python
observations
```

```python
Alcoholissue = tab.excel_ref('B3').expand(RIGHT).is_not_blank()
Alcoholissue
```

```python
Alcoholrelatedissue = tab.excel_ref('A5').expand(DOWN).is_not_blank()
Alcoholrelatedissue
```

```python
Dimensions = [
            HDim(Alcoholissue,'Alcohol related issue importance',DIRECTLY,ABOVE),
            HDim(Alcoholrelatedissue,'Alcohol related issue1',DIRECTLY,LEFT),
            HDimConst('Measure Type', 'Percent'),
            HDimConst('Unit','People')
            ]
```

```python
c1 = ConversionSegment(observations, Dimensions, processTIMEUNIT=True)
# savepreviewhtml(c1)
```

```python
new_table = c1.topandas()
new_table
```

```python
new_table.columns = ['Value' if x=='OBS' else x for x in new_table.columns]
```

```python
new_table['Value'] = pd.to_numeric(new_table['Value'], errors='coerce')
```

```python
new_table['Value'] = new_table['Value'].apply(str)
```

```python
def user_perc(x,y):
    
    if str(x) == 'Unweighted base':
        return 'Count'
    else:
        return y
    
new_table['Measure Type'] = new_table.apply(lambda row: user_perc(row['Alcohol related issue1'],row['Measure Type']), axis = 1)
```

```python
import io
import requests
url="https://raw.githubusercontent.com/ONS-OpenData/ref_alcohol/master/codelists/alcohol-related-issue.csv"
s=requests.get(url).content
c=pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
Final_table = pd.merge(new_table, c, how = 'left', left_on = 'Alcohol related issue1', right_on = 'Label')
```

```python
Final_table.columns = ['Alcohol related issue' if x=='Notation' else x for x in Final_table.columns]
```

```python
Final_table['Alcohol related issue importance'] = Final_table['Alcohol related issue importance'].map(
    lambda x: {
        'Primary issue' : 'primary', 
        'Secondary issue' : 'secondary',
        'Tertiary issue': 'tertiary' ,
        'Quaternary issue': 'quaternary',
        'Quinary issue': 'quinary'
        }.get(x, x))
```

```python
Final_table = Final_table[['Alcohol related issue','Alcohol related issue importance','Measure Type','Value','Unit']]
```
